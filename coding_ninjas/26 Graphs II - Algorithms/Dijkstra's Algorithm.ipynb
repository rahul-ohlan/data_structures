```python
# Given an undirected, connected and weighted graph G(V, E) 
# with V number of vertices (which are numbered from 0 to V-1) and E number of edges.
# Find and print the shortest distance from the source vertex (i.e. Vertex 0) 
# to all other vertices (including source vertex also) using Dijkstra's Algorithm.
```

```python
import queue
import sys
class Graph:
    
    def __init__(self,nVertices):
        
        self.nVertices = nVertices
        self.adjMatrix = [[0 for j in range(nVertices)] for i in range(nVertices)]
        
    def addEdge(self,v1,v2,wt):
        
        self.adjMatrix[v1][v2] = wt
        self.adjMatrix[v2][v1] = wt
        
    def removeEdge(self,v1,v2):
        if self.containsEdge is False:
            return
        self.adjMatrix[v1][v2] = 0
        self.adjMatrix[v2][v1] = 0
        
    def containsEdge(self,v1,v2):
        return True if self.adjMatrix[v1][v2] > 0 else False
    
    def __dfsHelper(self,v1,visited):
        
        print(v1)
        visited[v1] = True
        for i in range(self.nVertices):
            if self.adjMatrix[v1][i] > 0 and visited[i] > 0:
                self.__dfsHelper(i,visited)
    
    def dfs(self):
        visited = [False for i in range(self.nVertices)]
        for i in range(self.nVertices):
            if visited[i] == False:
                self.__dfsHelper(i,visited)
    
    def __bfsHelper(self,v1,visited):
        
        que = queue.Queue()
        que.put(v1)
        
        while not que.empty():
            curr = que.get()
            print(curr)
            for i in range(self.nVertices):
                if self.adjMatrix[curr][i] > 0 and visited[i] is False:
                    que.put(i)
                    visited[i] = True
                    
        return
        
    
    def bfs(self):
        visited = [False for i in range(self.nVertices)]
        for i in range(self.nVertices):
            self.__bfsHelper(i,visited)
    
    def __getMinVertex(self,distance,visited):
        
        min_vertex = -1
        for i in range(self.nVertices):
            if visited[i] is False and (min_vertex == -1 or distance[i] < distance[min_vertex]):
                min_vertex = i
                
        return min_vertex
        # TRY DOING THE ABOVE FUNCTION USING PRIORITY QUEUES    
        
    def dijkstra(self):
        
        visited = [False for i in range(self.nVertices)]
        distance = [sys.maxsize for i in range(self.nVertices)]
        distance[0] = 0
        
        for i in range(self.nVertices-1):
            
            min_dist_vertex = self.__getMinVertex(distance,visited)
            visited[min_dist_vertex] = True
            # now explore the neighbours of this vertex
            for j in range(self.nVertices):
                if self.adjMatrix[min_dist_vertex][j] > 0 and visited[j] is False:
                    dist = distance[min_dist_vertex] + self.adjMatrix[min_dist_vertex][j]
                    if distance[j] > dist:
                        distance[j] = dist
            # TRY TO DO THE ABOVE USING ADJACENCY LISTS INSTEAD OF ADJACENCY MATRIX
                        
        for i in range(self.nVertices):
            
            print(i, distance[i])
            
        return 
```
