```python
class Edge:
    
    def __init__(self,src,dest,wt):
        
        self.src = src
        self.dest = dest
        self.wt = wt
        
    def __lt__(self,other):
        
        return self.wt < other.wt
    
class Node:
    
    def __init__(self):
        
        self.parent = -1
        self.rank = 0
        
class Kruskal:
    
    def __init__(self,nVertices):
        
        self.nVertices = nVertices
        self.edgeMat = list()
        
    def addEdge(self,edge):
        
        self.edgeMat.append(edge)
        self.edgeMat.sort()
        
    def __getParent(self,v1,parentMat): # find using path compression
        
        if parentMat[v1].parent == -1:
            return v1
        
        parentMat[v1].parent = self.__getParent(parentMat[v1].parent,parentMat) # path compression
        
        return parentMat[v1].parent
        
    
    def unionfind(self):
        
        final_list = list()
        parentMat = [Node() for i in range(self.nVertices)]
        count = 0
        while count < self.nVertices -1:
            for edge in self.edgeMat:

                p1 = self.__getParent(edge.src,parentMat)
                p2 = self.__getParent(edge.dest,parentMat)

                r1 = parentMat[p1].rank
                r2 = parentMat[p2].rank

                if p1 != p2:

                    if r1 > r2:
                        parentMat[p2].parent = p1     # union using ranks: to reduce the height after union
                                                                # lower rank should always point to higher rank parent
                    elif r1 < r2:
                        parentMat[p1].parent = p2    # union based on ranks
                    else:
                        parentMat[p1].parent = p2     # union using ranks
                        parentMat[p2].rank += 1

                    count +=1
                    final_list.append(edge)
                    
        return final_list
        
```

```python
nVertices,nEdges = map(int,input().split())
rahul = Kruskal(nVertices)
for i in range(nEdges):
    
    v1,v2,wt = map(int,input().split())
    edge = Edge(v1,v2,wt)
    rahul.addEdge(edge)

res = rahul.unionfind()
for item in res:
    
    print(str(item.src) + " " + str(item.dest) + " " + str(item.wt))
```
