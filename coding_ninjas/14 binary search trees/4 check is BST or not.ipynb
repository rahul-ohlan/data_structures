```python
import queue
class binarytree:
    def __init__(self,data):
        self.data = data
        self.left = None
        self.right = None
        
def takeinput():
    qu = queue.Queue()
    print('enter root')
    rootdata = int(input())
    if rootdata!=-1:
        root = binarytree(rootdata)
    qu.put(root)
    while not qu.empty():
        currentnode = qu.get()
        print('enter left child of',currentnode.data)
        leftchilddata =int(input())
        if leftchilddata!=-1:
            leftchild = binarytree(leftchilddata)
            currentnode.left = leftchild
            qu.put(leftchild)
        print('enter right child of',currentnode.data)
        rightchilddata =int(input())
        if rightchilddata!=-1:
            rightchild = binarytree(rightchilddata)
            currentnode.right = rightchild
            qu.put(rightchild)
    return root

def printnicely(root):
    if root==None:
        return
    qu = queue.Queue()
    qu.put(root)
    while not qu.empty():
        current = qu.get()
        print(current.data,end=':')
        if current.left!=None:
            print('L',current.left.data,end=',')
            qu.put(current.left)
        else:
            print('L',-1,end=',')
        if current.right!=None:
            print('R',current.right.data)
            qu.put(current.right)
        else:
            print('R',-1)
    return

        
        
def leveltraversal(root):
    if root == None:
        return
    qu = queue.Queue()
    qu.put(root)
    qu.put('NULL')
    while not qu.empty():
        current = qu.get()
        if current!='NULL':
            print(current.data,end=' ')
            if current.left!=None:
                qu.put(current.left)
            if current.right!=None:
                qu.put(current.right)
        else:
            if qu.empty():
                return
            else:
                qu.put('NULL')
                print()
    return
```

```python
def mintree(root):
    if root==None:
        return 100000
    leftmin = mintree(root.left)
    rightmin = mintree(root.right)
    
    return min(leftmin,rightmin,root.data)
```

```python
def maxtree(root):
    if root == None:
        return -100000
    leftmax = maxtree(root.left)
    rightmax = maxtree(root.right)
    
    return max(leftmax,rightmax,root.data)
```

```python
def isBST(root):
    if root is None:
        return True
    leftmax = maxtree(root.left)
    rightmin = mintree(root.right)
    
    if root.data > rightmin or root.data< leftmax:
        return False
    isleftbst = isBST(root.left)
    isrightbst = isBST(root.right)
    return isleftbst and isrightbst
```

```python
# more optimized solution to check if tree is BST above is O(n**2)
def isBST2(root):
    if root is None:
        return 100000,-100000,True
    leftmin,leftmax,isleftbst = isBST2(root.left)
    rightmin,rightmax,isrightbst = isBST2(root.right)
    
    minimum = min(leftmin,rightmin,root.data)
    maximum = max(leftmax,rightmax,root.data)
    
    istreebst = True
    
    if root.data <= leftmax or root.data > rightmin:
        istreebst = False
    if not isleftbst or not isrightbst:
        istreebst = False
    return minimum,maximum,istreebst
```

```python
def isbst3(root,min_range,max_range):
    if root is None:
        return True
    if root.data<min_range or root.data>max_range:
        return False
    isleftwithinconstraints = isbst3(root.left,min_range,root.data-1)
    isrightwithinconstraints = isbst3(root.right,root.data,max_range)
    
    return isleftwithinconstraints and isrightwithinconstraints
```
