```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None

        
def printLL(head):
    while head is not None:
        print(str(head.data) + '-->', end= '')
        head = head.next
    print('none')

# Take two pointers, slow and fast: fast moves two times slow; so, when fast reaches the end of 
#  the list, slow will reach the mid point of the list
# in case of odd length of list: while fast.next is not None
# in case of the even length of the list: while fast.next.next is not None
def midpoint(head):
    slow = head
    fast = head
    while fast.next is not None and fast.next.next is not None:
        slow = slow.next
        fast = fast.next.next
    return slow
    
    
def mergesortedLL(head1,head2):
    if head1 is None:
        return head2
    if head2 is None:
        return head1
    fh = None
    ft = None
    if head1.data<=head2.data:
        fh = head1
        ft = head1
        head1 = head1.next
    else:
        fh = head2
        ft = head2
        head2 = head2.next
        
    while head1 is not None and head2 is not None:
        if head1.data <=head2.data:
            ft.next = head1
            ft = ft.next
            head1 = head1.next
        else:
            ft.next = head2
            ft = ft.next
            head2 = head2.next
    while head1 is not None:
        ft.next = head1
        ft = ft.next
        head1 = head1.next
        
    while head2 is not None:
        ft.next = head2
        ft = ft.next
        head2 = head2.next
        
    return fh
    
    
    
    
def reverseIteratively(head):
    prev = None
    current = head
    while current is not None:
        nxt = current.next
        current.next = prev
        prev = current
        current = nxt
    head = prev
    return head
        
        
def takeinput():
    inputList = [int(x) for x in input().split()]
    head = None
    tail = None
    for item in inputList:
        if item == -1:
            break
        newNode = Node(item)
        if head is None:
            head = newNode
            tail = newNode
        else:
            tail.next = newNode
            tail = newNode
    return head
```

```python
rahul = takeinput()
rahul2 = takeinput()
newhead = mergesortedLL(rahul,rahul2)
printLL(newhead)
```
