```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next = None
        pass
    
def findNodeRecursively(head,n):
    
    if head is None:
        return -1
    
    if head.data == n:
        return 0
    
    index = 1 + findNodeRecursively(head.next,n)
#  The Catch is to return -1 if index is zero, since index will become zero automatically 
#  if the element is not found in the list and first "if" loop returns -1 to previous call, making
#  the index to be zero. From here, we don't want index, we only want -1. 
    if index == 0:
        return -1
    else:
        return index
    
    
def printLL(head):
    while head is not None:
        print(head.data,end =' ')
        head = head.next
    print()
    
def evenafterodd(head):
    if head is None:
        return None
    oddlist = list()
    evenlist = list()
    counter = head
    while counter is not None:
        if counter.data ==0 or counter.data%2 ==0:
            evenlist.append(counter)
            counter = counter.next
        else:
            oddlist.append(counter)
            counter = counter.next
    head1 = None
    tail1 = None
    for item in oddlist:
        if head1 is None:
            head1 = item
            tail1 = item
        else:
            tail1.next = item
            tail1 = item
            tail1.next = None
    head2 = None
    tail2 = None
    for item in evenlist:
        if head2 is None:
            head2 = item
            tail2 = item
        else:
            tail2.next = item
            tail2 = item
            tail2.next = None

    if len(oddlist)>=1:
        tail1.next = head2
        return head1
    else:
        return head2


    
def takeinput():
    inputlist = [int(x) for x in input().split()]
    head = None
    tail = None
    for item in inputlist:
        if item == -1:
            break
        newnode = Node(item)
        if head is None:
            head = newnode
            tail = newnode
        else:
            tail.next = newnode
            tail = newnode
    return head

head = takeinput()
# index = findNodeRecursively(ruchika,5)
# print(index)
rahul = evenafterodd(head)
printLL(rahul)
```
