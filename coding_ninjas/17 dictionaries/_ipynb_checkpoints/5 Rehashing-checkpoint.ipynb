```python
class Mapnode:
    def __init__(self,key,value):
        self.key = key
        self.value = value
        self.next = None
```

```python
class Map:
    def __init__(self):
        self.bucketsize = 5
        self.buckets = [None for i in range(self.bucketsize)]
        self.count = 0
        
    def size(self):
        return self.count
    
    def getbucketindex(self,hashcode):
        
        return (abs(hashcode))%(self.bucketsize)
    
    def rehash(self):
        temp = self.buckets
        self.buckets = [None for i in range(2*self.bucketsize)]
        self.bucketsize = 2*self.bucketsize
        self.count = 0
        for head in temp:
            while head !=None:
                self.insert(head.key,head.value)
                head = head.next
    def LoadFactor(self):
        return self.count/self.bucketsize
    
    def insert(self,key,value):
        
        hc = hash(key)
        index = self.getbucketindex(hc)
        
        head = self.buckets[index]
        while head!=None:
            if head.key == key:
                head.value = value
                return
            head = head.next
        
        newnode = Mapnode(key,value)
#         newnode.next = head
        newnode.next = self.buckets[index]
        self.buckets[index] = newnode
        self.count+=1
        
        loadFactor = self.count/self.bucketsize
        if loadFactor >= 0.7:
            self.rehash()
        
    def searchmap(self,key):
        
        hc = hash(key)
        index = self.getbucketindex(hc)
        
        current = self.buckets[index]
        while current!=None:
            if current.key == key:
                print('key is present and value is:',current.value)
                return
            current = current.next
        print('key not found ')
            
    def deletenodeinmap(self,key):
        
        hc = hash(key)
        index = self.getbucketindex(hc)
        
        current = self.buckets[index]
        if current == None:
            return
        if current.key == key:
            self.buckets[index] = current.next
            self.count-=1
            return
        elif current.next==None:
            return
        while current!=None and current.next!=None:
            if current.next.key==key:
                current.next = current.next.next
                self.count-=1
                return
            current = current.next      
```
