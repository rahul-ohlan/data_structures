```python
class binarytree:
    def __init__(self,data):
        self.data = data
        self.left = None
        self.right = None
```

```python
class BST:
    
    def __init__(self):
        self.root = None
        self.numnodes = 0
        
        
        
    def insertHelper(self,root,data):
        if root == None:
            newnode = binarytree(data)
            root = newnode
            return root
        if data < root.data:
            root.left = self.insertHelper(root.left,data)
            return root
        else:
            root.right = self.insertHelper(root.right,data)
            return root
        
        
    def insert(self,data):
        self.numnodes+=1
        self.root = self.insertHelper(self.root,data)
        
    def ispresentHelper(self,root,data):
        if root == None:
            return False
        if data < root.data:
            return self.ispresentHelper(root.left,data)
        else:
            return self.ispresentHelper(root.right,data)
        
    def printbstHelper(self,root):
        if root == None:
            return
        print(root.data,end=':')
        if root.left!=None:
            print('L:',root.left.data,end=',')
        if root.right!=None:
            print('R:',root.right.data)
        else:
            print()
        self.printbstHelper(root.left)
        self.printbstHelper(root.right)
        
    def mini(self,root):
        if root == None:
            return 1000000
        leftmin = self.mini(root.left)
        rightmin = self.mini(root.right)
        
        return min(leftmin,rightmin,root.data)
        
        
    def deleteHelper(self,root,data):
        if root is None:
            return False,None
        if data < root.data:
            isdeleted,leftroot = self.deleteHelper(root.left,data)
            root.left = leftroot
            return isdeleted,root
        elif data >root.data:
            isdeleted,rightroot = self.deleteHelper(root.right,data)
            root.right = rightroot
            return isdeleted,root
        else:
            if root.left is None and root.right is None:
                return True,None
            
            if root.left is None:
                return True,root.right
            
            if root.right is None:
                return True,root.left
            
            else:
                replacement = self.mini(root.right)
                root.data = replacement
                isdeleted,newroot = self.deleteHelper(root.right,replacement)
                root.right = newroot
                return True,root
            
        
    def delete(self,data):
        isdeleted,newroot = self.deleteHelper(self.root,data)
        if isdeleted:
            self.numnodes-=1
        self.root = newroot
        return isdeleted
        
    def printbst(self):
        self.printbstHelper(self.root)
        
        
        
    def ispresent(self,data):
        return self.ispresentHelper(self.root,data)
        
        

    
    def count(self):
        return self.numnodes
```
