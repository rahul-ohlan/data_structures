```python
class BST:
    
    def __init__(self):
        self.root = None
        self.numnodes = 0
        
        
    def insertHelper(self,root,data):
        if root==None:
            newnode = binarytree(data)
            return newnode
        if data < root.data:
            root.left = self.insertHelper(root.left,data)
            return root
        else:
            root.right = self.insertHelper(root.right,data)
            return root
        
        
    def insert(self,data):
        self.numnodes+=1
        self.root = self.insertHelper(self.root,data)
        
        
    def mini(self,root):
        if root is None:
            return 100000
        leftmin = self.mini(root.left)
        rightmin = self.mini(root.right)
        
        return min(leftmin,rightmin,root.data)
        
    def deleteHelper(self,root,data):
        if root is None:
            return False,None
        if data < root.data:
            # root won't get changed in this case
            # return original root and make root.left = newleftroot 
            #(incase it gets changed in left subtree)
            
            isleftdeleted,leftroot = self.deleteHelper(root.left,data)
            root.left = leftroot
            return isleftdeleted,root
        
        if data > root.data:
            # since the root won't change in this case, return the original root
            # remember to make root.right = newrightroot(if it gets changed in right subtree)
            isrightdeleted,rightroot = self.deleteHelper(root.right,data)
            root.right = rightroot
            return isrightdeleted,root
        
        else:
            # be very careful here and condider all possibilities
            # 1. The root is a leaf:
            if root.left is None and root.right is None:
                return True,None
            #2. The root has only one child
            
            if root.right is None:
                return True,root.left
            if root.left is None:
                return True,root.right
            
            # 3. ROOT HAS BOTH CHILDREN
            # in case left side as two identical values and are maximum in the left subtree
            # it would fail the bst condition of all elements on left subtree < root
            # so in this case always take the minimum of right subtree as newroot
            # another thing is that we are not really deleting the root, but replacing root.data
            # -- with the min of right subtree and deleting the min from right subtree. 
            else:
                replacement = self.mini(root.right)
                root.data = replacement
                # delete the REPLACEMENT FROM THE THE RIGHT SUBTREE
                deleted,newrightnode = self.deleteHelper(root.right,replacement)
                print(newrightnode.left)
                root.right = newrightnode
                return True,root
            
            
            
        
    
    def delete(self,data):
        
        isdeleted,newroot = self.deleteHelper(self.root,data)
        if isdeleted:
            self.numnodes-=1
        self.root = newroot
        return isdeleted
    
    def count(self):
        return self.numnodes
    
    def searchHelper(self,root,data):
        if root is None:
            return False
        if data==root.data:
            return True
        if data < root.data:
            # call on left
            return self.searchHelper(root.left,data)
        else:
            # call on right side
            return self.searchHelper(root.right,data)
    
    
    
    def searchbst(self,data):
        return self.searchHelper(self.root,data)
    
    def printtreeHelper(self,root):
        if root==None:
            return
        print(root.data,end=':')
        if root.left!=None:
            print('L',root.left.data,end=',')
        if root.right!=None:
            print('R',root.right.data)
        else:
            print()
        self.printtreeHelper(root.left)
        self.printtreeHelper(root.right)
    
    
    def printtree(self):
        self.printtreeHelper(self.root)
```
