```python
class priorityQnode:
    def __init__(self,value,priority):
        self.value = value
        self.priority = priority
```

```python
# just for recap, Minheap:
class priorityQueue:
    
    def __init__(self):
        self.pq = list()
        
        
    def getsize(self):
        return len(self.pq)
        
    def isEmpty(self):
        return self.getsize ==0
    
    def __perlocateUp(self):
        
        childindex = self.getsize()-1
        while childindex > 0:
            parentindex = (childindex-1)//2
            if self.pq[childindex].priority < self.pq[parentindex].priority:
                self.pq[childindex],self.pq[parentindex] = self.pq[parentindex],self.pq[childindex]
                childindex = parentindex
            else:
                break
        pass
    def insert(self,value,priority):
        newnode = priorityQnode(value,priority)
        self.pq.append(newnode)
        self.__perlocateUp()
        pass
    def getMin(self):
        if self.isEmpty():
            return None
        return self.pq[0]

    def __perlocateDown(self):
        parentindex = 0
        child1 = 2*parentindex+1
        child2 = 2*parentindex+2
        Min = child1
        while child1 <= self.getsize()-1:
            if child2 <= self.getsize()-1 and self.pq[child2].priority < self.pq[child1].priority:
                Min = child2
            if self.pq[parentindex].priority > self.pq[Min].priority:
                self.pq[parentindex],self.pq[Min] = self.pq[Min],self.pq[parentindex]
                parentindex = Min
                child1 = 2*parentindex +1
                child2 = 2* parentindex+2
                Min = child1
            else:
                break
    
    def removeMin(self):
        if self.isEmpty():
            return None
        res = self.pq[0]
        self.pq[0] = self.pq[self.getsize()-1]
        self.pq.pop()
        self.__perlocateDown()
        
        return res.priority
        
    
```
