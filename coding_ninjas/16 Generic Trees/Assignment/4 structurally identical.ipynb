```python
import queue
class treenode:
    def __init__(self,data):
        self.data = data
        self.children = list()
        
def takeinput():
    print('enter root data')
    root_data = int(input())
    if root_data == -1:
        return None
    root = treenode(root_data)
    que = queue.Queue()
    que.put(root)
    while not que.empty():
        currentnode = que.get()
        print('enter the number of children of:',currentnode.data)
        childcount = int(input())
        for i in range(childcount):
            print('enter next child of:',currentnode.data)
            child_data = int(input())
            child = treenode(child_data)
            currentnode.children.append(child)
            que.put(child)
    return root

def printleveltree(root):
    if root is None:
        return
    que = queue.Queue()
    que.put(root)
    while not que.empty():
        current = que.get()
        print(current.data,end=':')
        for child in current.children:
            if child!= current.children[-1]:
                print(child.data,end=',')
                que.put(child)
            else:
                print(child.data,end='')
                que.put(child)
        print()
        
    return
```

```python
def isIdentical(root1,root2):
    if not root1 or not root2:
        return False
    Flag = True
    if root1.data==root2.data:
        if len(root1.children)==len(root2.children):
            for i in range(len(root1.children)):
                if root1.children[i].data!=root2.children[i].data:
                    Flag = False
                    return Flag
        else:
            return False
    else:
        return False
    # if we reach this line it means, Flag is True and trees are identical
    # considering first level children of respective roots
    for i in range(len(root1.children)):
        Flag = isIdentical(root1.children[i],root2.children[i])
        if Flag==False:
            return Flag
    return Flag
```

```python
root1 = takeinput()
root2 = takeinput()
print(isIdentical(root1,root2))
```
