```python
import queue
class binarytree:
    def __init__(self,data):
        self.data = data
        self.left = None
        self.right = None
def takelevelinput():
    q = queue.Queue()
    print('please enter the root value: ')
    rootdata = int(input())
    if rootdata == -1:
        return
    root = binarytree(rootdata)
    q.put(root)
    while not q.empty():
        currentnode = q.get()
        print('enter left child of:',currentnode.data)
        leftdata = int(input())
        if leftdata!=-1:
            leftchild = binarytree(leftdata)
            currentnode.left = leftchild
            q.put(leftchild)
        print('enter the right child of:',currentnode.data)
        rightdata = int(input())
        if rightdata!= -1:
            rightchild = binarytree(rightdata)
            currentnode.right = rightchild
            q.put(rightchild)
    return root

def printlevelwise(root):
    ruchika = queue.Queue()
    if root is None:
        return
    ruchika.put(root)
    while not ruchika.empty():
        currentnode = ruchika.get()
        print(currentnode.data,end=':')
        if currentnode.left is not None:
            print('L'+':',currentnode.left.data,end=',')
            ruchika.put(currentnode.left)
        else:
            print('L'+':',-1,end=',')
        if currentnode.right is not None:
            print('R'+':',currentnode.right.data)
            ruchika.put(currentnode.right)
        else:
            print('R'+':',-1)
    return

ruchikA = takelevelinput()
printlevelwise(ruchikA)
```
