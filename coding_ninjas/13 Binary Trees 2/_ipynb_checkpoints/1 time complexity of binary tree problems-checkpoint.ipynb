```python
class binarytree:
    def __init__(self,data):
        self.data = data
        self.left = None
        self.right = None
        
        
def takeinput():
    rootdata = int(input())
    if rootdata == -1:
        return None
    root = binarytree(rootdata)
    leftree = takeinput()
    rightree = takeinput()
    root.left = leftree
    root.right = rightree
    return root

def printnicely(root):
    if root == None:
        return
    print(root.data,end=': ')
    if root.left != None:
        print('L',root.left.data,end=',')
    if root.right!=None:
        print('R',root.right.data,end=' ')
    print()
    printnicely(root.left)
    printnicely(root.right)
    
def numnodes(root):
    if root == None:
        return 0
    left = numnodes(root.left)
    right = numnodes(root.right)
    return left + right + 1

def heightoftree(root):
    if root is None:
        return 0
    left = heightoftree(root.left)
    right = heightoftree(root.right)
    if left>=right:
        return left+1
    else:
        return right +1
    
def nodesatdepthK(root,K):
    if root==None:
        return 0
    if K==0:
        return 1
    left = nodesatdepthK(root.left,K-1)
    right = nodesatdepthK(root.right,K-1)
    return left+right

rahul = takeinput()
printnicely(rahul)
print('Number of Nodes in given tree are:',numnodes(rahul))
print('height of tree is:',heightoftree(rahul))
K= int(input('enter depth: '))
print('number of nodes at depth',K,'is:',nodesatdepthK(rahul,K))
```
